```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30, 25, 20, 13, 10, 2]
labels =['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
explode = [0.02] * 6
```

```python
plt.pie(values, labels=labels, colors=colors, explode=explode, autopct='%.1f%%')
plt.legend(loc=(1, 0.5))
plt.show()
```

```python
plt.pie(values, labels=labels, colors=colors, 
        explode=explode, autopct='%.1f%%', 
        startangle=90, counterclock=True,
        wedgeprops={'width':0.8})
plt.show()
```

```python
plt.pie(values, labels=labels, colors=colors,
        autopct='%.1f%%',
        startangle=90, counterclock=False,
        wedgeprops={'width':0.8, 'edgecolor':'w', 'linewidth':2})
plt.show()
```

```python
def custom_autopct(pct):
    return '{:.0f}%'.format(pct) if pct >=10 else ''
```

```python
plt.pie(values, labels=labels, colors=colors,
        autopct=custom_autopct, pctdistance=0.6, 
        startangle=90, counterclock=False,
        wedgeprops={'width':0.8, 'edgecolor':'w', 'linewidth':1})
plt.savefig('data/pie.png')
plt.show()
```

```python
df_stu = pd.read_csv('data/학생.csv')
df_stu
```

```python
df_pro = pd.read_csv('data/교수.csv')
df_pro
```

```python
df_merge = pd.merge(df_stu, df_pro, how='inner', left_on='지도교수', right_on='교수번호')
df_merge
```

```python
df_group = df_merge.groupby(['지도교수','교수이름']).size()
df_group
```

```python
labels = [f'{name[1]}({name[0]})' for name in df_group.index]
values = df_group.values    
```

```python
plt.pie(values, labels=labels, colors=colors,
        autopct='%.1f%%', startangle=90, counterclock=False,
        wedgeprops=wedgeprops)

plt.rc('font', size=20)
plt.title('교수별 학생 현황')
plt.show()
```

```python
df_group = df_stu.groupby('학생학과').size()
```

```python
labels = list(df_group.index)
values = df_group.values
values
```

```python
wedgeprops={'width':0.7, 'ec':'yellow', 'lw':3}
plt.title('학과별 학생수 현황')
plt.pie(values, labels=labels, colors=colors,
        autopct='%.1f%%', explode=[e*0.03 for e in values],
        startangle=90, counterclock=False,
        wedgeprops=wedgeprops)
plt.show()
```

```python
df_cou = pd.read_csv('data/강좌.csv')
df_cou
```

```python
df_enrol = pd.read_csv('data/수강.csv')
df_enrol
```

```python
def count_cou(code):
    filt = df_enrol['강좌번호'] == code
    df = df_enrol[filt]
    return int(df['강좌번호'].count())
```

```python
df_cou['수강인원'] = df_cou['강좌번호'].apply(count_cou)
df_cou
```

```python
df = df_cou[['강좌이름', '수강인원']]
df
```

```python
labels = df['강좌이름']
values = df['수강인원']
```

```python
plt.pie(values, labels=labels, colors=colors,
        startangle=90, counterclock=False,
        autopct='%.1f%%', 
        explode=explode, wedgeprops=wedgeprops)
plt.show()
```
