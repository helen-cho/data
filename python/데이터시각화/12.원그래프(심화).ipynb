```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30, 25, 20, 13, 10, 2]
labels =['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
explode = [0.02] * 6
```

```python
plt.pie(values, labels=labels, colors=colors, explode=explode, autopct='%.1f%%')
plt.legend(loc=(1, 0.5))
plt.show()
```

```python
plt.pie(values, labels=labels, colors=colors, 
        explode=explode, autopct='%.1f%%', 
        startangle=90, counterclock=True,
        wedgeprops={'width':0.8})
plt.show()
```

```python
plt.pie(values, labels=labels, colors=colors,
        autopct='%.1f%%',
        startangle=90, counterclock=False,
        wedgeprops={'width':0.8, 'edgecolor':'w', 'linewidth':2})
plt.show()
```

```python
def custom_autopct(pct):
    return '{:.0f}%'.format(pct) if pct >=10 else ''
```

```python
plt.pie(values, labels=labels, colors=colors,
        autopct=custom_autopct, pctdistance=0.6, 
        startangle=90, counterclock=False,
        wedgeprops={'width':0.8, 'edgecolor':'w', 'linewidth':1})
plt.savefig('data/pie.png')
plt.show()
```
