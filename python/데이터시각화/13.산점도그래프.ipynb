```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../데이터분석/data/score.xlsx')
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000  #np.random.rand(8) 0~1사이 랜덤 값 8개 발생
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 500
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수') #cmap='plasma', cmap='inferno', cmap='magma', cmap='cividis'
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3) #alpha 겹칠 경우
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5, orientation='horizontal')
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
names = df['이름']
names
```

```python
year = df['학년']
year
```

```python
x = df['국어']
x
```

```python
y = df['영어']
y
```

```python
plt.scatter(df['국어'], df['영어'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('국어 점수')
plt.ylabel('영어 점수')
for idx, name in enumerate(names):
    plt.text(x[idx], y[idx], str(year[idx]) + '.' + name, size=10)
```

```python
x = df['수학']
y = df['과학']
```

```python
plt.scatter(df['수학'], df['과학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('수학 점수')
plt.ylabel('과학 점수')
plt.xlim(0, 110)
plt.ylim(0, 110)
for idx, name in enumerate(names):
    plt.text(x[idx], y[idx], str(year[idx]) + '.' + name, size=10)
```
