```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.show()
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000  #np.random.rand(8) 0~1사이 랜덤 값 8개 발생
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.show()
```

```python
sizes = df['학년'] * 500
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.show()
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수') #cmap='plasma', cmap='inferno', cmap='magma', cmap='cividis'
plt.ylabel('수학 점수')
plt.show()
```

```python
year = list(df['학년'])
x = list(df['국어'])
y = list(df['영어'])
names = list(df['이름'])

plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=sizes, c=year, cmap='viridis', alpha=0.3)
plt.xlabel('국어 점수')
plt.ylabel('영어 점수')
plt.ylim(0, 110)
plt.xlim(0, 110)
for idx, name in enumerate(names):
    plt.text(x[idx], y[idx], f'{year[idx]}.{name}', size=7, ha='center')
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.8, orientation='horizontal')
plt.show()    
```
