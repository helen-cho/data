```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data/score.csv')
df
```

```python
labels = df['이름'][:3]
values = df['키'][:3]
```

```python
plt.bar(labels, values)
plt.show()
```

```python
colors = ['r', 'g', 'blue']
plt.bar(labels, values, color=colors, alpha=0.5)
plt.ylim(165, 200)
plt.show()
```

```python
plt.bar(labels, values, width=0.5)
plt.show()
```

```python
xticks = [x+'님' for x in labels]
yticks = [y for y in range(0, 250, 30)]
yticks1 = [str(y) + 'cm' for y in range(0, 250, 30)]
plt.bar(labels, values)
plt.xticks(labels, xticks, rotation=45)
plt.yticks(yticks, yticks1, rotation=45)
plt.show()
```
