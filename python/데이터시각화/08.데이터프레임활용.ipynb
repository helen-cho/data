```python
cols=df1.columns
cols
```

```python
matplotlib.rcParams['font.size'] = 10
plt.figure(figsize=(10, 5))
for i in range(1, len(cols), 1): #국어, 영어, 수학
    plt.plot(df['이름'], df[cols[i]], 'o-', lw=0.5, label=cols[i], ms=2)
    for j, score in enumerate(df[cols[i]]):
        plt.text(j, score+1, score, ha='center')

plt.legend(ncols=3, loc='lower left')
plt.show()
```

```python
def graph(no):
    filt = df.index == no
    df2 = df[filt][['국어', '영어', '수학', '과학', '사회']]
    
    if df2.shape[0] == 0:
        print(no + ' 학생이 없습니다.')
        return
  
    df2 = df2.iloc[0, :]
    name = df.loc[no, '이름']

    plt.title(name + '성적표')
    plt.bar(df2.index, df2.values)
    
    for i, value in enumerate(df2.values):
        plt.text(i, value+1, value)

    plt.show()
```

```python
graph('2번')
```
