```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title('Line Graph')
plt.show()
```

```python
plt.rc('font', family='Malgun Gothic', size=10)
plt.plot(x, y)
plt.title('꺽은선 그래프')
plt.show()
```
