```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus']= False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.title('꺽은선 그래프')
plt.show()
```

```python
plt.plot(x, y)
plt.title('꺽은선 그래프')
plt.xlabel('X축', color='red')
plt.ylabel('Y축', color='blue')
plt.show()
```

```python
plt.plot(x, y)
plt.title('꺽은선 그래프')
plt.xlabel('X축', color='red')
plt.ylabel('Y축', color='blue')
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```
