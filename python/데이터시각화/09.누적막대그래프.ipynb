```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data/score.csv', index_col='지원번호')
df
```

```python
plt.bar(df['이름'], df['국어'])
plt.bar(df['이름'], df['영어'])
plt.show()
```

```python
plt.bar(df['이름'], df['국어'])
plt.bar(df['이름'], df['영어'], bottom=df['국어'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(df['이름'], df['국어'])
plt.bar(df['이름'], df['영어'], bottom=df['국어'])
plt.bar(df['이름'], df['수학'], bottom=df['국어'] + df['영어'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(df['이름'], df['국어'], label='국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], label='영어')
plt.bar(df['이름'], df['수학'], bottom=df['국어'] + df['영어'], label='수학')

plt.xticks(rotation=60)
plt.ylim(10, 300)
plt.ylabel('합계')

for idx, name in enumerate(df['이름']):
    sum =  df.iloc[idx, 3] + df.iloc[idx, 4] + df.iloc[idx, 5]
    plt.text(idx, sum+1, sum, ha='center', color='blue')
plt.legend(loc=(1.01, 0.5))
plt.show()
```
