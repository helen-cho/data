```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
data = {
    '영화' : ['명량', '극한 직업', '신과 함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번 방의 선물', '암살'],
    '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270],  #(단위 : 만 명)
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
plt.bar(df['영화'], df['평점'])
plt.xticks(rotation=60)
plt.show()
```

```python
plt.bar(df['영화'], df['평점'])
plt.title('국내 Top8 영화 평점 정보')
plt.xticks(rotation=60)
plt.xlabel('영화')
plt.ylabel('평점')
plt.show()
```

```python
df_group = df.groupby('개봉 연도')[['관객 수','평점']].mean()
df_group
```

```python
plt.plot(df_group.index, df_group['평점'])
plt.show()
```

```python
plt.plot(df_group.index, df_group['평점'], marker='o')
plt.xticks([2005, 2010, 2015, 2020])
plt.ylim(7, 10)
```

```python
max = df['평점'].max()
min = df['평점'].min()
plt.plot(df_group.index, df_group['평점'], marker='o')
plt.xticks([i for i in range(2005, 2025, 5)])
plt.ylim(df['평점'].min()-0.2, df['평점'].max()+0.2)
plt.show()
```

```python
max = df['평점'].max()
min = df['평점'].min()
plt.plot(df_group.index, df_group['평점'], marker='o')
plt.xticks([i for i in range(2005, 2025, 3)])
plt.ylim(df['평점'].min()-0.2, df['평점'].max()+0.2)

year = list(df_group.index)
rating = list(df_group['평점'])
for idx, people in enumerate(df_group['관객 수']):
    print(idx, year[idx], rating[idx], people)
    plt.text(year[idx], rating[idx], str(f'{int(people):,}명'), size=10, color='red')
plt.show()
```

```python
filt = df['평점'] >= 9.0
df[filt]
df[~filt]
values = [len(df[filt]), len(df[~filt])]
labels = ['9점 이상', '9점 미만']

wedgeprops={'width':0.8, 'edgecolor':'w', 'linewidth':5}
plt.pie(values, labels=labels, autopct='%.1f%%', wedgeprops=wedgeprops)
plt.legend(loc=(1, 0.3))
plt.show()
```

```python
df.sort_values('개봉 연도', inplace=True)
```

```python
plt.figure(figsize=(10, 5))
plt.barh(df['영화'], df['관객 수'], color='orange', alpha=0.9)
plt.xlim(0, 2200)
movies = list(df['영화'])
for idx, people in enumerate(df['관객 수']):
    plt.text(250, movies[idx], str(f'{people:,}명({year[idx]}년)'), size=12, color='yellow')
plt.show()
```
