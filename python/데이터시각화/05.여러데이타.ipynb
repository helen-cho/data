```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
days = [1, 2, 3]  #1일, 2일, 3일
az = [2, 4, 8]  #[단위 : 만명] 1일부터 3일까지 아스트라제네카 접종 인구
pfizer = [5, 1, 3]  #화이자
moderna = [1, 2, 5]  #모더나
plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
plt.show()
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', linestyle='--')
plt.plot(days, moderna, label='moderna', marker='s', ls='-.')
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data/score.csv',  index_col='지원번호')
df
```

```python
df.loc['1번', '학교'] = '산북고'
df.loc['6번', '학교'] = '남능고'
```

```python
score = df[['이름', '국어', '영어', '수학']][:5]
score
```

```python
plt.figure(figsize=(12, 7))
matplotlib.rcParams['font.size'] = 10

cols = score.columns
for i in range(1, len(cols)): #국어, 영어, 수학
    plt.plot(score['이름'], score[cols[i]], 'o-', label=cols[i], alpha=0.7)

    for j, s in enumerate(score[cols[i]]):
        plt.text(score.iloc[j, 0], s+1.2, s, ha='center')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
matplotlib.rcParams['font.size'] = 10

plt.plot(score['이름'], score['국어'], 'o-', label='국어')
plt.plot(score['이름'], score['영어'], 's-', label='영어')
plt.plot(score['이름'], score['수학'], '*-', label='수학')
plt.legend()

for i, s in enumerate(score['국어']):
    plt.text(score.iloc[i, 0], s+2, s, ha='center')

for i, s in enumerate(score['영어']):
    plt.text(score.iloc[i, 0], s+1, s, ha='center')

for i, s in enumerate(score['수학']):
    plt.text(score.iloc[i, 0], s+1, s, ha='center')    

plt.show()
```

```python
score = df.groupby('학교')[['국어', '영어','수학']].mean()
score
```

```python
matplotlib.rcParams['font.size'] = 10
plt.figure(figsize=(12, 7))

cols = score.columns
for i in range(0, 3): #국어, 영어, 수학
    plt.plot(score.index, score[cols[i]], 'o-', label=cols[i])

    for j, s in enumerate(score[cols[i]]):
        plt.text(score.index[j], s+1, s, ha='center')
plt.legend()
plt.show()
```

```python
yticks = [i for i in range(50, 101, 5)]
matplotlib.rcParams['font.size'] = 10
plt.figure(figsize=(12, 7))

plt.plot(score.index, score['국어'], 'o-', label='국어')
plt.plot(score.index, score['영어'], 'o-', label='영어')
plt.plot(score.index, score['수학'], 'o-', label='수학')

plt.yticks(yticks)
plt.legend(loc=(1.01, 0.5))

for i, s in enumerate(score['국어']):
    plt.text(score.index[i], s+1, s, ha='center')

for i, s in enumerate(score['영어']):
    plt.text(score.index[i], s+1, s, ha='center')    

for i, s in enumerate(score['수학']):
    plt.text(score.index[i], s+1, s, ha='center')        
plt.show()
```
