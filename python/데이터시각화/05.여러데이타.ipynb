```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
days = [1, 2, 3]  #1일, 2일, 3일
az = [2, 4, 8]  #[단위 : 만명] 1일부터 3일까지 아스트라제네카 접종 인구
pfizer = [5, 1, 3]  #화이자
moderna = [1, 2, 5]  #모더나
plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
plt.show()
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', linestyle='--')
plt.plot(days, moderna, label='moderna', marker='s', ls='-.')
plt.legend()
plt.show()
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', linestyle='--')
plt.plot(days, moderna, label='moderna', marker='s', ls='-.')
plt.legend(ncol=3)
plt.show()
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', linestyle='--')
plt.plot(days, moderna, label='moderna', marker='s', ls='-.')
plt.legend(ncol=3)
plt.xticks([1, 2, 3])
plt.yticks([2, 4, 6, 8])
plt.xlabel('일', loc='right')
plt.ylabel('단위:만명', loc='top')
plt.show()
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', linestyle='--')
plt.plot(days, moderna, label='moderna', marker='s', ls='-.')
plt.legend(ncol=3)
plt.xticks([1, 2, 3])
plt.yticks([2, 4, 6, 8])
plt.xlabel('일', loc='right')
plt.ylabel('단위:만명', loc='top')

for idx, txt in enumerate(az):
    plt.text(days[idx], az[idx]+0.2, txt, ha='center', color='blue')

for idx, txt in enumerate(pfizer):
    plt.text(days[idx], pfizer[idx]+0.2, txt, ha='center', color='red')

for idx, txt in enumerate(moderna):
    plt.text(days[idx], moderna[idx]+0.2, txt, ha='center', color='green')   

plt.savefig('../shot.png', dpi=100)    
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data/score.csv',  index_col='지원번호')
df
```

```python
df.loc['1번', '학교'] = '산북고'
df.loc['6번', '학교'] = '남능고'
```

```python
score = df[['이름', '국어', '영어', '수학', '과학', '사회']][:5]
score
```

```python
plt.plot(score['이름'], score['국어'], 'o-', ms=10, label='국어')
plt.plot(score['이름'], score['영어'], 's-', label='영어')
plt.plot(score['이름'], score['수학'], '*-', label='수학')
plt.plot(score['이름'], score['과학'], 'x-', label='과학')
plt.plot(score['이름'], score['사회'], 'v:', label='사회')
plt.legend(loc=(1.05, 0.5))
plt.show()
```

```python
score = df.groupby('학교')[['국어', '영어','수학', '사회', '과학']].mean()
score
```

```python
yticks = [i for i in range(50, 101, 5)]
matplotlib.rcParams['font.size'] = 10
plt.figure(figsize=(10, 5))
plt.plot(score.index, score['국어'], 'o-', label='국어')
plt.plot(score.index, score['영어'], 'o-', label='영어')
plt.plot(score.index, score['수학'], 'o-', label='수학')
plt.plot(score.index, score['사회'], 'o-', label='사회')
plt.plot(score.index, score['과학'], 'o-', label='과학')
plt.yticks(yticks)
plt.legend(loc=(1.01, 0.5))
plt.show()
```
