```python
def merge_enrol_stu(code):
    filt = df_enrol['강좌번호'] == code
    df = df_enrol[filt]
    df_scode_sname = df_stu[['학생번호', '학생이름']]
    df_merge = pd.merge(df, df_scode_sname, how='inner', on='학생번호')
    return df_merge
```
