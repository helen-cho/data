```python
df_cou = pd.read_csv('강좌.csv')
df_cou
```

```python
#다음은 강좌번호를 입력받아 수강 신청한 학생 수를 구하는 함수이다. 밑줄에 알맞은 명령어를 입력하시오.(25~27)
def count_stu(code):
    filt = df_enrol['강좌번호'] == code
    df = df_enrol[filt]
    count = df['학생번호'].count()
    return count
```

```python
#28) df_cou '학생수'열에 위 함수를 적용하여 학생수를 구하시오. (2점)
df_cou['학생수'] = df_enrol['강좌번호'].apply(count_stu)
```

```python
df_cou
```

```python
#다음은 강좌번호를 입력받아 과목평균을 구하는 함수이다. 밑줄에 알맞은 명령어를 입력하시오. (28 ~ 29)
def avg_cou(code):
    filt = df_enrol['강좌번호'] == code
    df = df_enrol[filt]
    avg = df['점수'].mean()
    return avg
```

```python
df_cou['평균'] = df_enrol['강좌번호'].apply(avg_cou)
```
