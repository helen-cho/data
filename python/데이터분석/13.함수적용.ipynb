```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')
df
```

```python
def add_cm(height):
    if isinstance(height, int):
        return str(height) + 'cm'
    else:
        return int(height.replace('cm', ''))
```

```python
df['키'] = df['키'].apply(add_cm)
```

```python
df
```

```python
df['키'] = df['키'].apply(lambda x:str(x) + 'cm' if isinstance(x, int) else int(x.replace('cm', '')))
```

```python
df
```

```python
def capitalize(lang):
    if pd.notnull(lang):
        return lang.capitalize()
    else:
        return '없음'
```

```python
df['SW특기'] = df['SW특기'].apply(capitalize)
```

```python
df
```

```python
import numpy as np
```

```python
df['SW특기'] = df['SW특기'].apply(lambda x:x.capitalize() if x!='없음' else np.nan)
```

```python
df
```

```python
df_score = df.loc[:, '국어':'사회']
```

```python
df_score
```

```python
df_score['총점'] = df_score.apply(lambda x:sum(x['국어':'사회']), axis=1)
```

```python
df_score
```

```python
df_score['평균'] = df_score['총점']/5
```

```python
df_score
```

```python
def grade(avg):
    if avg>=90:
        return 'A'
    elif avg>=80:
        return 'B'
    elif avg>=70:
        return 'C'
    elif avg>=60:
        return 'D'
    else:
        return 'F'
    
```

```python
df_score['학점'] = df_score['평균'].apply(grade)
```

```python
df_score['학점'] = df_score.apply(lambda x:grade(x['평균']), axis=1)
```

```python
df_score
```

```python
import numpy as np
```

```python
df_score.loc['합계'] = df_score.apply(np.sum, axis=0)
df_score
```

```python
df_score.drop(index='합계', inplace=True)
```

```python
df_score.loc['과목평균'] = df_score.apply(np.mean, axis=0)
```
