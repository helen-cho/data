```python
import pandas as pd
df = pd.read_excel('data/score.xlsx', index_col='지원번호')
df
```

```python
def add_cm(height):
    if isinstance(height, int):
        return str(height) + 'cm'
    else:
        return int(height.replace('cm', ''))

df['키'] = df['키'].apply(add_cm)
df.head(5)
```

```python
df['키'] = df['키'].apply(lambda x:str(x) + 'cm' if isinstance(x, int) else int(x.replace('cm', '')))
df.head(5)
```

```python
def capitalize(lang):
    if pd.notnull(lang):
        return lang.capitalize()
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```

```python
def capitalize(lang):
    if pd.notnull(lang):
        return lang.capitalize()
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df.tail(4)
```

```python
df['SW특기'] = df['SW특기'].apply(lambda x: x.capitalize() if pd.notna(x) else x)
df.tail(4)
```

```python
df_score = df.loc[:, '국어':'사회']
df_score.head()
```

```python
df_score['총점']=df_score.apply(lambda row:row['국어']+row['영어']+row['수학']+row['과학']+row['사회'], axis=1)
df_score
```

```python
df_score['평균'] = df_score.apply(lambda row:row['총점']/5, axis=1)
df_score
```

```python
import numpy as np
df_score.loc['합계'] = df_score.apply(np.sum, axis=0)
df_score
```
