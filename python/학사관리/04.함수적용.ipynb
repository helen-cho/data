```python
import pandas as pd
df_enrol = pd.read_csv('data/수강.csv')
```

```python
df_enrol.fillna({'점수':0}, inplace=True)
```

```python
filt = df_enrol['강좌번호'] == 'E221'
df = df_enrol[filt]
count = df['점수'].count()
sum = df['점수'].sum()
print(f'합계:{int(sum)}점\n수강인원:{count}명')
```

```python
def sum(code):
    filt = df_enrol['강좌번호'] == code
    df = df_enrol[filt]
    sum = df['점수'].sum()
    return int(sum)
```

```python
print(f'합계:{sum('E221')}')
```

```python
def count_lcode(code):
    filt = df_enrol['학생번호'] == code
    df = df_enrol[filt]
    count = df['점수'].count()
    return int(count)
```

```python
df_stu = pd.read_csv('data/학생.csv')
```

```python
df_stu['수강과목수'] = df_stu['학생번호'].apply(count_lcode)
```

```python
def sum_lcode(code):
    filt = df_enrol['학생번호'] == code
    df = df_enrol[filt]
    sum = df['점수'].sum()
    return int(sum)
```

```python
df_stu['합계'] = df_stu['학생번호'].apply(sum_lcode)
```

```python
df_stu['평균'] = df_stu['합계'] / df_stu['수강과목수']
```
