```python
def merge_enrol_stu(code):
    filt = df_enrol['강좌번호'] == code
    df = df_enrol[filt]
    df_merge = pd.merge(df, df_stu, how='inner', on='학생번호')
    return df_merge
```

```python
df_merge = merge_enrol_stu('E221')
df_merge
```
