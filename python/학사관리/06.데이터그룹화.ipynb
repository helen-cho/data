```python
def count(code):
    filt = df['학생번호']==code
    if df[filt].empty:
        return 0
    else:
        return df[filt]['강좌번호'].count()
```

```python
df_stu['수강과목수'] = df_stu['학생번호'].apply(count)
```

```python
def mean(code):
    filt = df['학생번호']==code
    if df[filt].empty:
        return 0
    else:
        return df[filt]['점수'].mean()
```

```python
df_stu['평균'] = df_stu['학생번호'].apply(mean)
```

```python
def sum(code):
    filt = df['학생번호'] == code
    if df[filt].empty:
        return 0
    else:
        return df[filt]['점수'].sum()
```

```python
df_stu['합계'] = df_stu['학생번호'].apply(sum)
```

```python
def count_cou(code):
    filt = df['강좌번호'] == code
    if df[filt].empty:
        return 0
    else:
        return df[filt]['학생번호'].count()
```

```python
df_cou['수강인원수'] = df_cou['강좌번호'].apply(count_cou)
```

```python
def mean_cou(code):
    filt = df['강좌번호'] == code
    if df[filt].empty:
        return 0
    else:
        return df[filt]['점수'].mean()
```

```python
df_cou['평균'] = df_cou['강좌번호'].apply(mean_cou)
```

```python
def sum_cou(code):
    filt = df['강좌번호'] == code
    if df[filt].empty: return 0
    else:
        return df[filt]['점수'].sum()
```

```python
df_cou['합계'] = df_cou['강좌번호'].apply(sum_cou)
```
